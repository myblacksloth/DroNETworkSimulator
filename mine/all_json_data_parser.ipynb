# json parser

caricare i file json nella directory "./tuttidati/" prima di procedere

N.B.: al termine della simulazione in Colab tutti i file vengono eliminati

librerie

```python
import json
from os import listdir
from os.path import isfile, isdir, join
from tqdm import tqdm
import matplotlib.pyplot as plt
```

utilizzo un dizionario per memorizzare tutti i dati che mi servono

differenzio i dati come segue:

numero di droni / algoritmo /delivery ratio

```python
retrieved = {}
for i in range(100): # per i droni da 0 a 100
    retrieved[i] = {
        'RoutingAlgorithm.QL': None,
        "RoutingAlgorithm.RND" : None,
        "RoutingAlgorithm.GEO" : None
    }
print(retrieved)
```

leggo tutti i file json presenti nella cartella dei dati

```python
path = "./tuttidati/"
files = [join(path, f) for f in listdir(path) if isfile(join(path, f))]
print(f"{len(files)} files to read...")
```

leggo i vari file

```python
for cur_file in tqdm(files, desc="lettura file..."):
    with open(cur_file, 'r') as f:
        dati = json.load(f)
        algoritmo = dati['mission_setup']['routing_algorithm']
        numero_droni = dati['mission_setup']['n_drones']
        delivery_ratio = dati['packet_delivery_ratio']
        retrieved[numero_droni][algoritmo] = delivery_ratio
```

```python
for i in tqdm(range(100), desc="ottengo i dati..."):
    if retrieved[i]['RoutingAlgorithm.QL'] and retrieved[i]['RoutingAlgorithm.RND']:
        dati_ql.append(retrieved[i]['RoutingAlgorithm.QL'])
        dati_rnd.append(retrieved[i]['RoutingAlgorithm.RND'])
        x_axis.append(i)
```

visualizzo grafico dei dati

```python
# plt.figure(figsize=(10,10))
plt.plot(dati_ql, color='r', label='qmr')
plt.plot(dati_rnd, color='g', label='random')
# plt.plot(dati_geo, color='b', label='qgeo')
plt.legend(loc="upper left") # mostrare la legenda in alto a sinistra
plt.grid() # mostra la griglia
```

```python
# plt.figure(figsize=(10,10))
plt.title('QMR vs RND')
plt.xlabel('Numero di droni')
plt.ylabel('Packet Delivery Ratio')
plt.plot(x_axis, dati_ql, color='r', marker='o', linestyle='--', label='qmr')
plt.plot(x_axis, dati_rnd, color='g', marker='o', linestyle='--', label='random')
# plt.plot(dati_geo, color='b', label='qgeo')
plt.legend(loc="upper left") # mostrare la legenda in alto a sinistra
plt.grid() # mostra la griglia
```
